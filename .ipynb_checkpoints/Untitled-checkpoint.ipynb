```python
%matplotlib inline
import numpy as np
from numpy.linalg import pinv,norm
from numpy.random import multivariate_normal, shuffle
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt


is_shuffle = True
```

# Part 1

```python
# number of samples Done
n1,n2 = 200,200
# +1 dataset configuration
mean1, cov1 = np.array((1,1)), np.array(((3,2),(2,3)))
# -1 dataset configuration
mean2, cov2 = np.array((-1,-1)), np.array(((2,-1),(-1,2)))
```

```python
# data generation Done
dataset1 = multivariate_normal(mean1, cov1, n1)
label1 = np.ones(n1)
dataset2 = multivariate_normal(mean2, cov2, n2)
label2 = -1 * np.ones(n2)
```

```python
# training/testing dataset generation Done

# Dataset +1
shuffled_index1 = np.arange(n1)
shuffle(shuffled_index1)
training_data1, training_label1 = dataset1[shuffled_index1[:100]],label1[shuffled_index1[:100]] 
testing_data1, testing_label1 = dataset1[shuffled_index1[100:]],label1[shuffled_index1[100:]]

# Dataset -1
shuffled_index2 = np.arange(n2)
shuffle(shuffled_index2)
training_data2, training_label2 = dataset2[shuffled_index2[:100]],label2[shuffled_index2[:100]] 
testing_data2, testing_label2 = dataset2[shuffled_index2[100:]],label2[shuffled_index2[100:]]
```

```python
# data concatenation 
training_data = np.concatenate((training_data1, training_data2))
training_label = np.concatenate((training_label1, training_label2))
testing_data = np.concatenate((testing_data1, testing_data2))
testing_label = np.concatenate((testing_label1, testing_label2))

training_size = training_data.shape[0]
testing_size = testing_data.shape[0]

# is_shuffle (to be done)
#if (is_shuffle):
    
```

```python
# plot Done

# training data
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.scatter(training_data1[:,0],training_data1[:,1] , c='r', marker='o',s=10)
myplot.scatter(training_data2[:,0],training_data2[:,1] , c='b', marker='o',s=10)
plt.title("training data plot")
plt.show()

# testing data
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.scatter(testing_data1[:,0],testing_data1[:,1] , c='r', marker='o',s=10)
myplot.scatter(testing_data2[:,0],testing_data2[:,1] , c='b', marker='o',s=10)
plt.title("testing data plot")
plt.show()
```

# Part 2

```python
# mean Done
training_mean = np.mean(training_data,axis=0)
testing_mean = np.mean(testing_data,axis=0)

# covariance
training_cov = (training_data - training_mean).T.dot(training_data - training_mean)/(training_size-1)
testing_cov = (testing_data - testing_mean).T.dot(testing_data - testing_mean)/(testing_size-1)
```

```python
# REport the estimated covariance matrix Done
print(training_cov)
print(testing_cov)
```

```python
# LDA Done
pi_1 = training_data1.shape[0]/training_size
pi_2 = training_data2.shape[0]/training_size
x1_mean = np.mean(training_data1,axis=0)
x2_mean = np.mean(training_data2, axis=0)
cov_i = pinv(training_cov)
w = ( x1_mean - x2_mean ).T.dot(cov_i)
b = x1_mean.dot(cov_i).dot(x1_mean.T)/2 - x2_mean.dot(cov_i).dot(x2_mean.T)/2 + np.log(pi_1/pi_2)
```

```python
# test Done
y_predict = testing_data.dot(w)+b

# regulation
for idx, v in enumerate(y_predict):
    if v>0:
        y_predict[idx] = 1
    else:
        y_predict[idx] = -1
        
error_rate = np.sum(np.abs(y_predict- testing_label))/2/testing_size

print("Accuracy : %s%%" % (100-error_rate*100))
```

```python
# misclassification Done
comp_array = np.abs(y_predict- testing_label)
misc_idx = np.nonzero(comp_array)
misc_data = testing_data[misc_idx]

# plot
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.scatter(testing_data[:,0],testing_data[:,1] , c='b', marker='o',s=10, label='testing data')
myplot.scatter(misc_data[:,0],misc_data[:,1] , c='r', marker='o',s=10, label='misclassification')
plt.title("misclassification data plot")
plt.legend()
plt.show()
```

# Part 3

```python
# initializing w
w = np.zeros((1,training_data.shape[1]+1))

# concatenate ones on training_data
training_data_modified = np.concatenate((training_data, np.ones((training_size,1))), axis=1)
testing_data_modified = np.concatenate((testing_data, np.ones((testing_size,1))), axis=1)

converged = False

while not converged:
    w_hist = w

    p = np.exp(w.dot(training_data_modified.T))/(1+np.exp(w.dot(training_data_modified.T)))
    s = p*(1-p)
    z = w.dot(training_data_modified.T) + np.divide(training_label.reshape((-1,1)).T - p, 1e-4+s) 
    S = np.diag(s.reshape(-1))

    w = (pinv(training_data_modified.T.dot(S).dot(training_data_modified))).dot(training_data_modified.T).dot(S).dot(z.T).T
    w = w/norm(w)
    if np.isclose(w, w_hist, atol=1e-5).all():
        converged = True
    print(w)
    
```

```python
# test

y_predict = testing_data_modified.dot(w.T).reshape((-1))


# regulation
for idx, v in enumerate(y_predict):
    if v>0:
        y_predict[idx] = 1
    else:
        y_predict[idx] = -1

error_rate = np.sum(np.abs(y_predict- testing_label))/2/testing_size

print("Accuracy : %s%%" % (100-error_rate*100))
```
